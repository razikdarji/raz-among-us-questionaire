```python
import pandas as pd



import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'pokemon_data.csv'
pokemon = pd.read_csv(path)
```

```python
type(pokemon)
```

```python
pandas.core.frame.DataFrame
```

```python
plt.hist(pokemon.hp)
```

```python
pokemon_groupings = pokemon.groupby("type1")
```

```python
water_pokemon = pokemon_groupings.get_group('water')
```

```python
type(water_pokemon)
```

```python
plt.hist(water_pokemon.hp, bins=25, label="red")
```

```python
df_water = pokemon_groupings.get_group('water')
df_fire = pokemon_groupings.get_group('fire')

plt.hist(df_water.hp, bins=20, label='water')
plt.hist(df_fire.hp, bins=20, label='fire')

plt.legend()
plt.title("Distribution of HP by type")
plt.xlabel("HP")
plt.ylabel("Number of Pokemon")
```
