```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'data/Questionaire'
questionaire_data = pd.read_csv('data/Questionaire.csv')
```

```python
# To display the first few rows of the data
print(questionaire_data.head())

# To get a summary of your dataframe
print(questionaire_data.info())
```

```python
fav_teacher = questionaire_data['who is your favourite teacher? '].value_counts()

# Plotting
fav_teacher.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])  # You can choose different colors
plt.title('Favorite Teachers in CTA Class')
plt.xlabel('Teacher')
plt.ylabel('Number of Votes')

# Optional: Add annotations (numbers) on each bar for clarity
for index, value in enumerate(fav_teacher):
    plt.text(index, value, str(value))

plt.show()
```

```python
locked_in_hogarth = questionaire_data['Hogarth accidently locked you in, I ate all the food so theres none left in the whole building, you have to nominate one CTA or HEX person to be stuck with all night, who do you pick?'].value_counts()

# Plotting
locked_in_hogarth.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon', 'black', 'yellow', 'red', 'purple', 'orange', 'brown', 'pink', 'gray', 'cyan', 'navy', 'lime', 'maroon', 'olive'])  # Colors for each bar
plt.title('Preferred Companion for Being Locked in Overnight')
plt.xlabel('Person')
plt.ylabel('Number of Votes')

# Adding annotations on each bar
for index, value in enumerate(locked_in_hogarth):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```
